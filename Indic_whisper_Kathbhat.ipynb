```python
from whisper_jax import FlaxWhisperForConditionalGeneration, FlaxWhisperPipline
import jax.numpy as jnp

pipeline = FlaxWhisperPipline('parthiv11/indic_whisper_nodcil', dtype=jnp.bfloat16)
```

```python
# testing out the pipeline
transcript = pipeline('/content/drive/MyDrive/hindi/wavs/844424930324566-536-m.wav')
print(transcript)
```

```python
# Creating a dictionary with audio file and ground truth transcript from the manifest.json file
ground = {}
file = open('/content/drive/MyDrive/hindi/manifest.json','r')
for line in file:
    data = json.loads(line)
    path = data['audio_filepath']
    text = data['text']
    audio_name = path.split("/")[3]
    ground[audio_name] = text


for key,value in ground.items():
    print(key,value)
```

```python
#1. Looping over the WAV directory containing all the audio file
#2. Getting the audio file transripts from indic_whisper pipeline and storing it as key value pair txt file

index = 1
audio_dir = '/content/drive/MyDrive/hindi/wavs'
with open('/content/drive/MyDrive/transcripts.txt','w') as opt:
  for filename in os.listdir(audio_dir):
    audio_file = os.path.join(audio_dir,filename)
    transcript = pipeline(audio_file)
    audio_name = audio_file.split('/')[6]
    text = transcript['text']
    opt.write(f'{audio_name}={text}')
    print(f'{index}. {audio_name} = {text}')
    index += 1
```

```python
# Creating dict for transcripts
transcripts = {}

file = open('/content/drive/MyDrive/transcripts (1).txt','r')

for line in file:
  splits = line.split('=')
  audio_name = splits[0]
  transcript = splits[1][:-1]
  transcripts[audio_name] = transcript


print(len(transcripts))
```

```python
# Getting missing audio transcripts
audio_dir = '/content/drive/MyDrive/hindi/wavs'
write_list = []
index = 1
with open('/content/drive/MyDrive/transcripts (1).txt','a') as opt:
  for key,_ in ground.items():
    if key not in transcripts:
      path = os.path.join(audio_dir,key)
      transcript = pipeline(path)
      text = transcript['text']
      transcripts[key] = text
      write_list.append(f'{key}={text}\n')
      print(f'{key}={text}')

  opt.writelines(write_list)
```

```python
# Calculating WER
hyp = []
grndT = []

for key,value in ground.items():
  hyp.append(transcripts[key])
  grndT.append(value)


measures = jiwer.compute_measures(grndT, hyp)
print(measures['wer'])
```
